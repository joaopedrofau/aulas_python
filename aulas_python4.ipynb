```python
pares = 0

for i in range(1, 6):
    num = int(input(f"Digite o {i} número: "))
    
    if num % 2 == 0:
        pares += 1

print(f"Você digitou {pares} números pares.")
```

```python
usuario_correto = "JOAOPEDRO"
senha_correta = "4321"

tentativas = 3

while tentativas > 0:
    print(f"\nTentativas restantes: {tentativas}")
    user = input("Usuário: ")
    senha = input("Senha: ")

    if user == usuario_correto and senha == senha_correta:
        print("Login feito com sucesso! Seja Bem-vindo.")
        break  
    else:
        tentativas -= 1
        print("Usuário ou senha incorretos.")

if tentativas == 0:
    print("\nAcesso bloqueado! Você excedeu o limite de 3 tentativas.")
```
