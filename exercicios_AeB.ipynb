```python
#6
from collections import deque

fila = deque()

while True:
    print("----Menu do cartório----")
    print("1- Cadastrar seu nome")
    print("0- Sair")

    opcao = input("Escolha uma opção: ")

    if opcao == "1":
        nome = input("Digite seu nome: ")
        fila.append(nome)
        print("Nome cadastrado")

    elif opcao == "0":
        print("Programa encerrado")
        break

    else:
        print("Inválido")
```
