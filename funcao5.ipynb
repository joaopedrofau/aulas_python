```python

nome_completo = input("Digite seu nome completo: ")


nomes = nome_completo.split()

print(f"\nAnalisando o nome: {nome_completo}")
print("-" * 40)


for nome in nomes:
   
    tamanho = len(nome)
    
   
    posicao = nome_completo.find(nome)
    
   
    invertido = nome[::-1]
    
    print(f"Nome: {nome}")
    print(f"  - Caracteres: {tamanho}")
    print(f"  - Posição inicial: {posicao}")
    print(f"  - Invertido: {invertido}")
    print("-" * 20)
```

```python

def imprimir_pares():
    
    for numero in range(1, 21):
       
        if numero % 2 == 0:
            print(numero)


imprimir_pares()
```

```python

def converter_para_maiusculas(texto):
   
    return texto.upper()


resultado = converter_para_maiusculas("olá, João Pedro!")


print(resultado)
```

```python

def atualizar_lista(lista):
  
    if len(lista) == 4:
        # Adiciona 2 novos elementos à lista
        lista.append("Elemento 5")
        lista.append("Elemento 6")
        
       
        print(f"Lista atualizada: {lista}")
    else:
        print("A lista inicial deve conter exatamente 4 elementos.")


minha_lista = ["A", "B", "C", "D"]


atualizar_lista(minha_lista)
```

```python

def mostrar_elementos(obrigatorio, *extras):
    print(f"Argumento formal: {obrigatorio}")
    if extras:
        print(f"Elementos extras: {extras}")
    else:
        print("Nenhum elemento extra fornecido.")
    print("-" * 30)


mostrar_elementos("Primeiro")


mostrar_elementos("Início", "Item A", "Item B", "Item C")
```
